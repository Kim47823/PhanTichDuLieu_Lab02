```python
import pandas as pd


#Đọc và tải dữ liệu từ tệp CSV:
file_path = 'dulieuxettuyendaihoc.csv'
df = pd.read_csv(file_path)

#Tính toán TBM cho từng chỉ số từ 1 đến 3:
cols = ['T', 'L', 'H', 'S', 'V', 'X', 'D', 'N']
for i in range(1, 4):
    df[f'TBM{i}'] = df[[f'{col}{i}' for col in cols]].mean(axis=1)

#Hàm xếp loại dựa trên giá trị TBM:
def xep_loai(tbm):
    if tbm < 5.0:
        return 'Y'
    elif tbm < 6.5:
        return 'TB'
    elif tbm < 8.0:
        return 'K'
    elif tbm < 9.0:
        return 'G'
    return 'XS'

#Áp dụng hàm xếp loại vào từng cột TBM:
for i in range(1, 4):
    df[f'XL{i}'] = df[f'TBM{i}'].apply(xep_loai)

#Lưu DataFrame đã xử lý trở lại vào tệp CSV gốc:
df.to_csv(file_path, index=False)

#Thông báo hoàn thành:
print("Đã tạo và lưu các biến xếp loại thành công!")
```

```python
import pandas as pd

#Đọc và tải dữ liệu từ tệp CSV:
file_path = 'dulieuxettuyendaihoc.csv'
df = pd.read_csv(file_path)

#Tính toán TBM cho từng chỉ số từ 1 đến 3:
cols = ['T', 'L', 'H', 'S', 'V', 'X', 'D', 'N']
for i in range(1, 4):
    df[f'TBM{i}'] = df[[f'{col}{i}' for col in cols]].mean(axis=1)

#Hàm chuyển đổi điểm TBM sang điểm theo hệ thống điểm bên Mỹ:
def Chuyen_Sang_Diem_USA(tbm):
    return (tbm / 10) * 4

#Áp dụng hàm chuyển đổi vào từng cột TBM:
for i in range(1, 4):
    df[f'US_TBM{i}'] = df[f'TBM{i}'].apply(Chuyen_Sang_Diem_USA)

#Lưu DataFrame đã xử lý trở lại vào tệp CSV gốc:
df.to_csv(file_path, index=False)

#Thông báo hoàn thành và hiển thị mặc định 5 hàng đầu tiên
print("Đã chuyển đổi và lưu các điểm trung bình thành công!")
df.head()
```

```python
import pandas as pd

#Đọc và tải dữ liệu từ tệp CSV:
file_path = 'dulieuxettuyendaihoc.csv'
df = pd.read_csv(file_path)

#Định nghĩa hàm tính toán kết quả xét tuyển (KQXT)
def calculate_kqxt(row):
    group = row['KT']
    DH1, DH2, DH3 = row['T1'], row['L1'], row['H1']
    if group in ['A', 'A1']:
        average_score = (DH1 * 2 + DH2 + DH3) / 4
    elif group == 'B':
        average_score = (DH1 + DH2 * 2 + DH3) / 4
    else:
        average_score = (DH1 + DH2 + DH3) / 3
    return 1 if average_score >= 5.0 else 0

#Áp dụng hàm tính toán vào từng hàng
df['KQXT'] = df.apply(calculate_kqxt, axis=1)
#Lưu DataFrame đã xử lý trở lại vào tệp CSV gốc và hiển thị mặc định 5 hàng đầu tiên
df.to_csv(file_path, index=False)
print(df.head())
```
